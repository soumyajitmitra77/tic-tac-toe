```python
## libraries

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from dataset import load_data
```

```python
#data
X,Y = load_data()
print(X.shape, Y.shape)

#for i in range(len(X)):
 #   print(X[i] ,':', Y[i])
```

### Model

We build a model initially with 5 layers,  all with ReLu activation, except the last layer, for which we use sigmoid, i.e.,
$$ f(x) = 1/(1 + e^{-x}) $$

```python
model = Sequential(
        [
         Dense(60, activation='relu', name = 'layer1', input_shape = (9,)),
         Dense(50, activation='relu', name = 'layer2'),
         Dense(30, activation='relu', name = 'layer3'),
         Dense(15, activation='relu', name = 'layer4'),
         Dense(1, activation='sigmoid', name = 'layer5')
        ]
)

model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.001),
    metrics = ['accuracy']
)

model.fit(
    X, Y,
    epochs = 100,
)
```

### Predictions

Change the numbers in the '[  ]' to predict your game.

```python

X_test = np.array([
    [4, 6, 6, 6, 4, 6, 5, 6, 4], 
])

prediction = model.predict(X_test)

#print("predictions = \n", prediction)

yhat = np.round(prediction)
        
print(f"result = \n{yhat}")
```
